Process field "Price"

```python
new_df['Price ($)'] = df['Price ($)'].replace('HB ','')
# thay thế tất cả các chuỗi "HB " (có khoảng trắng ở cuối) bằng chuỗi rỗng trong cột Price, tức là xóa các khoảng trắng đó đi
```

```python
new_df['Price'] = df['Price ($)']
# tạo thêm một cột mới tên là "Price" trong DataFrame new_df, với dữ liệu được lấy từ cột "Price ($)"
```

```python
new_df = new_df.drop(['Price ($)'], axis=1)
# Câu lệnh trên được sử dụng để xóa cột "Price ($)" khỏi new_df
```

```python
new_df['Price'] = new_df['Price'].replace(['HB '],' ', regex=True )
# Câu lệnh này sử dụng phương thức replace() để thay thế giá trị "HB " bằng dấu cách (" ") trong cột "Price" của DataFrame new_df. 
# Tham số regex=True được sử dụng để áp dụng việc thay thế này cho tất cả các chuỗi ký tự chứa "HB " trong cột "Price"
```

```python
convert_dict = {'Price': int}
new_df = new_df.astype(convert_dict)
# Câu lệnh trên sử dụng phương thức astype() để chuyển đổi kiểu dữ liệu của cột "Price" 
# trong new_df từ kiểu dữ liệu ban đầu sang kiểu dữ liệu số nguyên (integer)
```

```python
new_df['Price (VND)'] = new_df['Price']*682
# tạo ra một cột mới trong new_df với tên là "Price (VND)" và chứa giá trị của cột "Price" nhân với 682 (tỷ giá VNĐ/USD)
new_df.drop(["Price"], axis=1)
# loại bỏ cột "Price" khỏi  new_df
```

Process field Travel Time

```python
new_df['Travel Time'] = new_df['Travel Time'].replace([' hr '],'_', regex=True )
new_df['Travel Time'] = new_df['Travel Time'].replace([' min'],'', regex=True )
new_df['Travel Time'] = new_df['Travel Time'].replace(['_'],':', regex=True )
new_df['Travel Time'] = new_df['Travel Time'].replace([' hr'],':00', regex=True )
# Dòng đầu tiên new_df['Travel Time'] = new_df['Travel Time'].replace([' hr '],'_', regex=True ) thay thế khoảng trắng trước và sau chuỗi "hr" bằng dấu gạch dưới "_" trong cột "Travel Time".
# Dòng thứ hai new_df['Travel Time'] = new_df['Travel Time'].replace([' min'],'', regex=True ) thay thế chuỗi "min" bằng chuỗi rỗng "" trong cột "Travel Time".
# Dòng thứ ba new_df['Travel Time'] = new_df['Travel Time'].replace(['_'],':', regex=True ) thay thế dấu gạch dưới "_" bằng dấu hai chấm ":" trong cột "Travel Time".
# Dòng cuối cùng new_df['Travel Time'] = new_df['Travel Time'].replace([' hr'],':00', regex=True ) thay thế chuỗi "hr" bằng chuỗi ":00" trong cột "Travel Time".
```

```python
new_df[(new_df['Travel Time'].str.find('min') == 1)]
# lọc các hàng trong new_df mà giá trị trong cột "Travel Time" có chứa chuỗi "min" ở vị trí thứ hai
# mục đích để kiểm tra còn hàng nào trong new_df có giá trị trong cột "Travel Time" chứa chuỗi "min" hay không
```

```python
new_df[(new_df['Travel Time'].str.find(':') == -1)]
# lọc các hàng trong new_df mà giá trị trong cột "Travel Time" không chứa dấu hai chấm ":"
```

```python
new_df.loc[new_df['Travel Time'].str.find(':') == -1, ['Travel Time']] = '0:' + new_df['Travel Time']
# thêm dấu hai chấm ":" vào giá trị trong cột "Travel Time" của các hàng trong new_df mà giá trị này không chứa dấu hai chấm ":"
```

```python
# chuyển đổi giá trị trong cột "Travel Time" từ định dạng giờ:phút sang đơn vị giờ
def convertToHours(row):
   timeStr = row['Travel Time']
   units = timeStr.split(":")
   hour = float(units[0])
   min = float(units[1])
   return hour + min/60
# áp dụng hàm này cho mỗi hàng trong new_df, lưu kết quả vào cột mới "Travel Hours
new_df['Travel Hours'] = new_df.apply(convertToHours, axis=1)
```

```python
new_df = new_df.drop(['Travel Time'], axis=1)
# xoá cột Travel Time
```

Process Depart Time

```python
# chuyển đổi giá trị trong cột "Depart Time (Leg 1)" từ định dạng giờ:phút AM/PM sang đơn vị giờ và đơn vị phút
def convertDepartTimeToHour(row):
   timeStr = row['Depart Time (Leg 1)'][0:-2]
   units = timeStr.split(":")
   hour = float(units[0])
   return hour

def convertDepartTimeToMin(row):
  timeStr = row['Depart Time (Leg 1)'][0:-2]
  units = timeStr.split(":")
  min = float(units[1])
  return min
```

```python
new_df['Depart Hour'] = new_df.apply(convertDepartTimeToHour, axis=1)
new_df['Depart Min'] = new_df.apply(convertDepartTimeToMin, axis=1)
# hêm hai cột mới vào new_df với tên là "Depart Hour" và "Depart Min". 
# Các giá trị trong cột "Depart Time (Leg 1)" được chuyển đổi sang đơn vị giờ và đơn vị phút tương ứng
```
