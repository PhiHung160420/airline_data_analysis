Process field "Price"

```python
new_df['Price ($)'] = df['Price ($)'].replace('HB ','')
# thay thế tất cả các chuỗi "HB " (có khoảng trắng ở cuối) bằng chuỗi rỗng trong cột Price, tức là xóa các khoảng trắng đó đi
```

```python
new_df['Price'] = df['Price ($)']
# tạo thêm một cột mới tên là "Price" trong DataFrame new_df, với dữ liệu được lấy từ cột "Price ($)"
```

```python
new_df = new_df.drop(['Price ($)'], axis=1)
# Câu lệnh trên được sử dụng để xóa cột "Price ($)" khỏi new_df
```

```python
new_df['Price'] = new_df['Price'].replace(['HB '],' ', regex=True )
# Câu lệnh này sử dụng phương thức replace() để thay thế giá trị "HB " bằng dấu cách (" ") trong cột "Price" của DataFrame new_df. 
# Tham số regex=True được sử dụng để áp dụng việc thay thế này cho tất cả các chuỗi ký tự chứa "HB " trong cột "Price"
```

```python
convert_dict = {'Price': int}
new_df = new_df.astype(convert_dict)
# Câu lệnh trên sử dụng phương thức astype() để chuyển đổi kiểu dữ liệu của cột "Price" 
# trong new_df từ kiểu dữ liệu ban đầu sang kiểu dữ liệu số nguyên (integer)
```

```python
new_df['Price (VND)'] = new_df['Price']*682
# tạo ra một cột mới trong new_df với tên là "Price (VND)" và chứa giá trị của cột "Price" nhân với 682 (tỷ giá VNĐ/USD)
new_df.drop(["Price"], axis=1)
# loại bỏ cột "Price" khỏi  new_df
```
